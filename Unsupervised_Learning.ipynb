## **1. Clustering**

1) 정의: 비슷한 x들끼리 군집을 형성하는 것

2) 두가지 Type

- **Hierarchical algorithms** (bottom-up) : 각 x가 하나의 cluster라고 두는 것에서 시작해서, 비슷한 것끼리 묶어나가는 과정

>> 묶어나가는 과정에서 **cluster사이의 거리**에 대한 정의가 필요함

- **Partitional algorithms** (top-down) : 모든 x가 하나의 cluster라고 두고, 비슷한 것끼리 쪼개나가는 과정

3) **K-Means** (대표적인 clustering 기법)

- 원리: Initalization 후 두가지 단계를 반복하며 이루어짐
> - **Initialization**: 처음에 랜덤으로 K개의 cluster center를 뿌림
> - **Step a**: 가장 가까운 cluster center가 있는 곳으로 data point에 대해 cluster를 지정
> - **Step b**: 그 cluster로 지정된 data point들의 평균으로 cluster center를 바꿈
