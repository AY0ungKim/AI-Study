## **1. Decision Trees**

 종류: Outcome variable에 따라 두가지로 나뉨

- Regression Tree (Outcome이 continuous)

- Classification Tree (Outcome이 category)

장단점:

- 장점: 해석력이 좋음

- 배경) Decision Tree에서는 만약에 해당 region에 관측값이 너무 작을 경우 그 값이 대푯값이 될 수 있기에, Overfitting의 위험성이 높음

>> -> Overfitting을 방지하기 위해서 Decision tree에서는 *모델을 간단하게 만들어야 = split을 적게 하여 더 작은 tree를 만들어야* 함

## **3. Classification Trees**


- categorical response를 예측할 때 쓰는 방식

- 가장 많이 속한 값을 가진 class를 대푯값으로 사용하면 됨

- 기본 방식은 regression 이랑 매우 비슷



**1) 수행 알고리즘**

Step 1: J개의 region으로 splitting하기

> recursive binary splitting을 수행

> **lowest classification error rate를 달성하는 것이 목표**

> 여기서는 RSS를 계산할 수 없기에, **각 region의 *Impurity*의 합이 최소화**되는 방향으로 진행함

> image.png


>> 분류 트리를 만들 때는, 각 class의 값들이 균등하게 분배되어 있으면 분류의 정확도가 떨어질 수 있으므로, 균등하지 않고 편차가 심한 데이터가 더 좋음 (Impurity가 높음 = 균등하게 분배되어 있음)

>> Impurity 계산하는 방식은 다음 두 가지 방식 활용

>> - **Entropy**

>>> 낮아지는 방향으로 진행해야 함

>>> 엔트로피가 높다 = 각 class의 데이터의 분포가 균등하다

>> - **Gini-Index**

>>> 낮을수록 좋은 것

>>> Entropy랑 혼용해도 상관없음

## **4. Decision Tree의 장단점**

- 장점: **해석력이 좋으**므로, 의사결정에 활용도가 높다. 그리고 시각화하기도 좋고, 질적인 변수들을 활용하기에 좋다

- 단점: **정확도가 비교적 떨어질** 수 있으며, **robust하지 않다**
