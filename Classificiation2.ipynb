## **1. Bayes Clssifier**

- Bayes Classifier는 *risk R(f)를 최소화*하는 가장 좋은 classifier
image.png

>> Risk란? Loss의 기대값

- Binary classification에서, y를 **{+1, -1}**으로 분류

>> log함수의 특성 (1보다 크면 양수, 1보다 작으면 음수)을 이용하여서,sign 함수 안의 부호를 판별하고,이 부호를 통해 최종으로 **+1** or **-1**을 판별함.  

>> image.png

- **베이즈 정리**를 활용하여, posterior probability인 *p(y=1Ix=x)* 값을 추정
> image.png
>-  **p(y)** = prior probability = x를 고려하지 않은 원래 y class의 확률
>-  **p(xIy)** = likelihood = class y에서 관측된 x들의 probability의 곱
>- 기본적으로 linear 가정으로 인해 분류한 logistic regression (**Discriminative classifier**)과는 달리, Bayes classifier는 베이즈 정리를 사용하여서 *새로 확률을 계산*해서 만들고, 이를 통해 분류한 것이기 때문에, **generative classifier**라고 칭함

## **3. Classifiaction Model 평가하기**

(예시)

image.png


- **Misclassification rate** = 틀리게 예측한 거/전체 = (23+252)/10000 = 2.75%

- **False positive rate** = *positive라고 예측*했는데, *False 틀린*경우 = 23/9667 = 0.2%

- **False negative rate** = *negative라고 예측*했는데, *False 틀린*경우 = 252/333 = 75.7%

> 위 예시는 no라고 예측할 때 yes인 확률이 너무 높음 -> p(Yes I x=x)를 0.5를 기준으로 판별을 하는데, 이 경우 **threshold를 조금 낮춰서** no라고 예측할 경우를 줄일 수 있음

image.png


> False negative rate를 줄이기 위해 threshold를 낮춰야 함.


- **ROC Curve** = false negative rate와 true positive rate의 관계를 시각화한 그래프로 Area Under the Curve (AUC)를 계산 (파란색 부분)하여 성능 평가
