## **2. Padding & Stride**


- **Padding**

> activation map이 점점 더 작아지는 것을 방지하기 위해 input image의 경계선에 **zero pad**를 추가하는 방식

> 사이즈가 동일하게 유지하게 하는 P = (F-1)/2 (이때, F=filter의 가로/세로 길이)


>> image.png


- **Stride**

> 고해상도 이미지에서 계산이 너무 복잡해질 때, 이를 방지하기 위해 **보폭을 크게** 설정하는 방법

>>- 만약 stride=2이라면, 보폭을 두칸씩 이동하면서 filter 계산

>>> image.png

- output size = (N - F + 2P) / stride +1

- parameter의 number : filter의 size와 동일 = filter의 개수 x (filter의 차원 + 1)

> fully-connected라고 가정하면, paramter의 개수는 ouput size x (input size + 1)이기 때문에 CNN을 활용하면 parameter의 개수를 줄일 수 있음

## **3. Pooling Layer**

- 역할: 네트워크의 파라미터 갯수나 연산량을 줄이기 위해 input에서 spatial 하게 downsampling을 진행해 사이즈를 줄이는 역할로, **정해진 연산을 반복적**으로 수행하는 것이 특징

-  Pooling(downsampling)이 필요한 이유는, featuremap의 weight parameter 갯수를 줄이기 위함

- hyperparameter: **stride**, **filter size**

- 종류
>- max layer (가장 흔하게 쓰임) : filter 내의 최대값으로 이용 (유의미한 정보가 높은 signal (activation function의 output)을 가질것이라는 가정)
>- average layer : filter내의 평균값으로 이용 (noise도 줄이고 그 영역을 잘 설명하는 값)
