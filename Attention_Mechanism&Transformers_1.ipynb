## 1. Attention Mechanism


- 배경: RNN에서 seq to seq처럼 encoder-decoder structure을 사용할 때는 마지막 hidden stage만 이용하는데, 이렇게 하면 너무 긴 문장을 다 기억하지 못할 수 있음. 그렇다면, decoding 시에 앞선 encoder의 부분들을 참고하면서 진행하면 더 잘되지 않을까?

- Attention Idea: **query (해당 decoder의 hidden stage)**에서, query와 **key(참고하는 encoder의 hidden stage들)** 의 관련성을 계산해서 그를 score로 이용하여 확률 변환 후, **value (encoder의 hidden stage들)**와 곱해서 더한 **attention value (weighted average of values)**를 구하는 방식

> 관련한 논문 읽을 때, key, value, query를 찾아가면서 읽으면 됨

<br/>

- **Dot-Product Attnetion**

> 1. decoder의 query 정의 (t step에서, h차원의 s_t라 가정)

> 2. T개의 encoder hidden state로 key와 query의 연관성 구하기 - **attention score**

>> image.png

> 3. softmax 함수를 이용해서 **attention coefficient**구하기 (확률처럼)

>> image.png

> 4. encoder hidden state **value**와 attention coefficent를 곱한 것의 합인 **weighted sum of the encoder hidden states**를 구해서, **attention value** 구하기.

>> image.png

> 5. 원래 query와 나열하고, 차원이 커지는 것을 방지하여 필요하다면 FC layer 하나를 통과시켜서 하나의 s_t를 만들고, 그것을 이용해 y_t+1를 추정하고, 그 값은 다음 query의 input으로 들어감 (auto-regressive)

## 2. Transformers


- 배경: encoder-decoder에서 encoder key들을 참조하는 attention model을 **self-attention**으로, 해당 문장을 중심 단어를 기준으로, 주변 단어들과의 relevance를 계산해서 weighted sum을 만들어서 context를 포함한 형태로 **transform**해보자!
